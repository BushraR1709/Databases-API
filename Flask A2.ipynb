Q1. Explain GET and POST methods.



```python
GET :
The GET method is one of the HTTP request methods used to request data from a specified resource, typically a web server.
In Flask, we can handle GET requests by defining a route with the @app.route decorator and specifying methods=['GET'].
In GET, data passes through URL.

POST :
The POST method is another HTTP request method used to submit data to be processed to a specified resource, typically a web server.
In Flask, you can handle POST requests by defining a route with the @app.route decorator and specifying methods=['POST'].
In POST, data passes through body.
```

Q2. Why is request used in Flask?



```python
In Flask, the request object is a built-in feature that allows you to access and manipulate incoming HTTP request data. 
It provides access to various parts of an HTTP request, such as headers, query parameters, form data, files, and more.
The request object is essential for handling and processing client requests within Flask application.
    
```

Q3. Why is redirect() used in Flask?



```python
In Flask, the redirect() function is used to perform a client-side redirection to a different URL.
It is a powerful tool for controlling the flow of web applications by guiding users from one route to another. 
```

Q4. What are templates in Flask? Why is the render_template() function used?



```python
Templates :
    
In Flask, templates are a way to separate the structure and presentation of web pages from the actual application logic.
Templates allow us to create HTML pages with placeholders for dynamic content that can be generated and filled in by our Flask application.



render_template() function : 
    
The render_template() function in Flask is used to render the templates.
It takes the name of the template file as an argument and optionally accepts variables that can be passed to the template for rendering
dynamic content.
render_template() is a crucial function in Flask because it enables us to create dynamic, maintainable, and user-friendly web pages by rendering HTML
templates with dynamic content. It promotes clean code organization and enhances the overall development experience in Flask applications.
```

```python
from flask import Flask, jsonify

app = Flask(__name__)

# Sample data (usually data would come from a database)
items = [
    {"id": 1, "name": "Item 1"},
    {"id": 2, "name": "Item 2"},
]

# Route to get a list of items
@app.route('/items', methods=['GET'])
def get_items():
    return jsonify(items)

if __name__ == '__main__':
    app.run(host="0.0.0.0")

```
