Q1. What is Flask Framework? What are the advantages of Flask Framework?



```python
Flask Framework :
A Flask is a Python library that helps to access the method or function into outer world through url.
Flask is often referred to as a "micro-framework" because it provides the basic tools and components
needed for web development.
Flask is used for building web applications using the Python programming language.
Flask is often chosen for its simplicity, making it an excellent option for rapid prototyping, 
small to medium-sized web applications, and educational purposes.

Advantages :
- Flask Framework is lightweight and minimalistic.
- It is easy to learn and use.
- Flask is suitable for building RESTful APIs, making it a popular choice for creating web services and microservices.
- Flask is well-suited for small to medium-sized web applications and microservices.
- Flask uses the Jinja2 template engine, which simplifies the process of rendering dynamic HTML content and allows for code reusability.
```

Q3. What is App routing in Flask? Why do we use app routes?


```python
In Flask, app routing refers to the process of associating URLs (Uniform Resource Locators) with specific functions in your web application. 
These routes define how your application responds to different HTTP requests (e.g., GET, POST) and specify what content or data should be served
when a particular URL is accessed. App routing is a fundamental concept in Flask and is accomplished using the @app.route decorator.

we use app routes for URL Mapping.
App routing allows us to map specific URLs to functions or views within our Flask application. When a user visits a URL, 
Flask uses routing to determine which function should handle the request and generate the appropriate response.
App routing in Flask is a crucial feature for creating web applications because it defines how your application responds to different
URLs and HTTP requests. It provides a clean and organized way to map URLs to view functions and handle user interactions effectively.
```

Q4.  Create a “/welcome” route to display the welcome message “Welcome to ABC Corporation” and a “/” 
route to show the following details:

    Company Name: ABC Corporation

    Location: India

    Contact Detail: 999-999-9999 


Attach the screenshot of the output in Jupyter Notebook.


```python
from flask import Flask

app = Flask(__name__)

@app.route("/welcome")
def welcome():
    return "<h1>Welcome to ABC Corporation</h1>"

@app.route("/")
def company_info():
    company_name = 'Company Name: ABC Corporation'
    location = 'Location: India'
    contact_detail = 'Contact Detail: 999-999-9999'

    return f"{company_name}<br>{location}<br>{contact_detail}"

if __name__ == '__main__':
    app.run(host = "0.0.0.0")
```

Q5. What function is used in Flask for URL Building? Write a Python code to demonstrate the working of the 
url_for() function.

```python
In Flask, the url_for() function is used for URL building. It generates URLs for routes based on the name of the route function. 
This is useful for creating links to other pages within your Flask application without hardcoding URLs.
It ensures that your URLs remain consistent even if you change the route patterns.
```

```python
from flask import Flask, url_for

app = Flask(__name__)

# Define a route named 'hello'
@app.route('/hello')
def hello():
    return 'Hello, World!'

# Define a route named 'greet' that takes a parameter 'name'
@app.route('/greet/<name>')
def greet(name):
    return f'Hello, {name}!'

if __name__ == '__main__':
    with app.test_request_context():
        # Using url_for to generate URLs
        hello_url = url_for('hello')
        greet_url = url_for('greet', name='John')

        print(f'URL for hello route: {hello_url}')
        print(f'URL for greet route: {greet_url}')
```
